```python
# args et kwargs
def doubler(f):
    """ doubler est une fonction d'ordre superieure, car prend en entree une autre fonction
    ceci ne marche plus lorsque la fonction f prend en entree deux argument et plus"""
    def g(x):
        return 2 * f(x)
    return g

def f1(x):
    return x+1

def f2(x,y):
    return x+y

g = doubler(f1)
print(g(3))
print(g(-1))


# pour palier a cela , on va utiliser le deballage d'arguments
def magic(*args, **kwargs):
    print("unnamed args:", args)
    print("keyword args", kwargs)
# args est un tuple de ses arguments anomymes et kwargs un dict de ses arguments nommes
magic(1,2,key="word", key2="word2")

def other_way_magic(x,y,z):
    return x+y+z

x_y_list = [1,2]
z_dict = {"z":3}
print(*x_y_list)
#print(**z_dict)
print(other_way_magic(*x_y_list,**z_dict))

def doubler_correct(f):
    """ fonction d'ordre superieur qui prend en entree une fonction f pouvant prendre des argume nts arbitraires"""
    def g(*args,**kwargs):
        return 2 * f(*args, **kwargs)
    return g
g = doubler_correct(f2)
print(g(1,2))
```

```python
# Les vecteurs
# La methode la plus simple consiste a reperesenter les vecteurs comme une liste de nombres.
def vector_add(v,w):
    """ ajoute les elements correspondants"""
    return[v_i  + w_i
          for v_i, w_i in zip(v,w)]

def vector_subtract(v,w):
    """ ajoute les elements correspondants"""
    return[v_i  - w_i
          for v_i, w_i in zip(v,w)]

def vector_sum(vectors):
    """ fait la somme de tous les elements correspondants"""
    result= vectors[0]
    for vector in vectors[1:]:
        result = vector_add(result, vector)
    return result


# En y reflechissant bien, nous sommes entrain de faire un reduce de la liste de vecteur
# a l'aide de la fonction vector_add, donc nous pouvons reecrire avec les fonctions d'ordre superieur

def vector_sum_sup(vectors):
    return reduce(vector_add, vectors)


def scalar_multiply(c,v):
    """ c est un nombre, v est un vecteur"""
    return [c * v_i for v_i in v]

def vector_mean(vectors):
    """Calcule le vecteur dont le ieme element est la moyenne des iemes elements
     des vecteurs d'entrees"""
    n = len(vectors)
    return scalar_multiply(1/n, vector_sum_sup(vectors))

def dot(v,w):
    """ Produit scalaire de deux vecteurs"""
    return sum(v_i*w_i
              for v_i, w_i in zip(v,w))

def sum_of_squares(v):
    """ somme des carres d'un vecteur"""
    return dot(v,v)

def magnitude(v):
    """ calcule de la longueur d'un vecteur"""
    return math.sqrt(sum_of_squares(v))

def squared_distance(v,w):
    """somme des differences des carres de deux vecteurs"""
    return sum_of_squares(vector_subtract(v,w))

def distance(v,w):
    return math.sqrt(squared_distance(v,w))


def distance_m(v,w):
    return magnitude(vector_subtract(v,w))
        
```

```python
# Les matrices
A = [[1,2,3],
    [4,5,6]]

B = [[1,2],
    [3,4],
    [5,6]]

def shape(A):
    num_rows = len(A)
    num_cols = len(A[0]) if A else 0  #nbre d'element de la premiere ligne
    return num_rows, num_cols

def get_row(A,i):
    return A[i]  # retourne la ligne i
    
def get_column(A,j):
    return [A_i[j]
           for A_i in A]  # retourne la colonne j


def make_matrix(num_rows,num_cols, entry_fn):
    """ retourne une matrice num_rows* num_cols dont l'entree de range(i,j) est entry)fn()"""
    return [[entry_fn(i,j)            # soit i, creer un liste
            for j in range(num_cols)] #[entry_fn(i,0),....]
           for i in range(num_rows)]   # creer une liste pouyr chaque i

def is_diagonal(i,j):
    """des 1 sur la diagonale, des o partout ailleurs"""
    return 1 if i==j else 0

def is_not_diagonal(i,j):
    return i if i != j else 0
    

        
```

```python
def de_mean(x):
    """calcul la difference entre chaque lement de x et la moyenne de x"""
    x_bar = stat.mean(x)
    return [x_i - x_bar for x_i in x]


def variance(x):
    """ calcul de la variance de x en supposant que x a aumoins 2 elements"""
    n = len(x)
    deviations = de_mean(x)
    print(deviations)
    print(sum_of_squares(deviations))
    return sum_of_squares(deviations) / (n - 1)


def standard_deviation(x):
    return math.sqrt(variance(x))


def covariance(x,y):
    n = len(x)
    return dot(de_mean(x), de_mean(y)) / (n - 1)


def correlation(x,y):
    stdev_x = standard_deviation(x)
    stdev_y = standard_deviation(y)
    if stdev_x > 0 and stdev_y > 0:
        return covariance(x, y) / (stdev_x * stdev_y)
    else:
        return 0
```

```python
y = [1,2,3]
z = [4,5,6]
print(de_mean(y))
print("la variance",variance(y))
print("l'ecart type",standard_deviation(y))
print("covariance", covariance(y,y))
print("correlation", correlation(y,z))
```
